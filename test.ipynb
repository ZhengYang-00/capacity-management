建模整理
1. 站点信息，输入一个人每个站点游客去向分布情况
2. 车辆信息，当前载客量，当前续航状况，当前所在站点（当前最近站点距离）
3. 道路拓扑信息，道路拥堵情况
4. 每辆车按周期分配任务，当一辆车结束当前周期任务以后，再分配其新的任务

优化目标
1. 车站滞留人数最小化，游客乘车时间最小化
2. 次要目标：车辆运行路程、时间最短



两种方案：
1. 按车进行调用，每辆车结束当前运输任务后，根据当前最需要车辆的路线运营。
2. 按整体进行调用，按某个固定周期进行调用，每次调用只考虑空闲的运营车辆。

采用方案2的算法逻辑：
算法周期开始：获取当前路况信息、可用车辆信息、站点游客信息。
路径规划分配，每辆车依次选择下一目标站点，设计一个合理的代价函数，考虑路程、拥堵情况、下一站点人数、已经过站点游客的目的地情况、车辆载客量……
每一辆车的预期行驶时间超出了周期则停止其迭代，所有车辆迭代结束则该算法周期结束
输出每辆车的路径规划（站点列表）

站点之间的路径规划和代价预测当作路径规划部分，根据路径和拓扑结构进行规划。

需要的数据：
静态信息：
    1. 园区范围内的道路拓扑结构，以何种形式展现
    2. 园区内的站台信息，站台的位置
动态信息：
    1. 道路的实时情况，如拥堵情况等
    2. 站台的实时游客数量以及游客目的地分布
    3. 车辆的实时信息，包括位置、剩余载客数量、续航等
    

周期开始时：
    获取每个站台的旅客数量，以及旅客的目标站台分布，通过表或者字典的形式发送：
    {
        Station_A: {'total': 20, 'A': 0, 'B': 10, 'C': 5, 'D': 5...}
        ...
        Station_N: {'total': 10, 'A': 2, 'B': 5, ... 'N': 0}
    }
    获取空闲的车辆子集：
        avail_bus = [bus if bus.is_avail == True for bus in bus_set]

    for bus in avail_bus:
        调用Choose_Next_Station函数为当前车辆选择下一站台
        检查车辆完成当前任务所需的时间，调用路径规划算法实现:
        if time_cost > Loop_time:
            bus.is_avail = Fasle
            avail_bus.delete(bus) //如果所需时间超过周期时长，则不再为其分配站点
        if avail_bus is empty:
            可分配任务车辆数量为0时，结束运输任务分配
    
    最终给出每辆车的运输任务，即站台行驶顺序：
    bus.station_list = ['B','A','D','K'...]


创建bus类：
    class bus:
        包括以下信息
        bus.is_avail: 车辆是否可以分配任务
        bus.position: 车辆当前位置，位置可以是在某条路上或者在某个站台
        bus.capacity: 车辆的剩余载客量
        bus.passenger: 车上乘客的目标站点分布，以字典形式给出{'A': 5, 'B': 5, 'C': 2, ...}
        bus.endurance: 车辆的续航

关键函数Choose_Next_Station:
    该函数的输入包括当前车辆以及其相关信息，当前各站台的乘客需求表
    获取车辆当前位置，通过轨迹规划算法计算出车辆到各站点的时间和路程
    代价函数考虑因素：
        到达目标站点所需时间，时间越多，代价函数值越大
        目标站点的乘客人数，人数越多，代价函数值越小
        已过站点的综合target站点乘客数分布，即将已经经过的站点在车辆经过时的乘客目标站点分布情况求和，人数越多的站点代价函数值越小
    选择代价函数最小的站点作为车辆的下一停靠站
